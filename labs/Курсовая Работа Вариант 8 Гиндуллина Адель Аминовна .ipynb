–í–æ –≤—Ç–æ—Ä–æ–º –ø—Ä–∏–º–µ—Ä–µ –æ–¥–∏–Ω –∏–∑ –≤–æ–∑–º–æ–∂–Ω—ã—Ö —Ö–æ—Ä–æ—à–∏—Ö –º–∞—Å—Å–∏–≤–æ–≤ ‚Äî [1,1,1‚éØ‚éØ,1‚éØ‚éØ] (–∑–∞–º–µ–Ω–µ–Ω–Ω—ã–µ —ç–ª–µ–º–µ–Ω—Ç—ã –ø–æ–¥—á–µ—Ä–∫–Ω—É—Ç—ã).

–í —Ç—Ä–µ—Ç—å–µ–º –ø—Ä–∏–º–µ—Ä–µ –º–∞—Å—Å–∏–≤ —Ö–æ—Ä–æ—à–∏–π –∏–∑–Ω–∞—á–∞–ª—å–Ω–æ–µ.

–í–æ —á–µ—Ç–≤–µ—Ä—Ç–æ–º –ø—Ä–∏–º–µ—Ä–µ –æ–¥–∏–Ω –∏–∑ –≤–æ–∑–º–æ–∂–Ω—ã—Ö —Ö–æ—Ä–æ—à–∏—Ö –º–∞—Å—Å–∏–≤–æ–≤ ‚Äî [‚àí2.5‚éØ‚éØ‚éØ‚éØ‚éØ‚éØ‚éØ‚éØ,‚àí2,‚àí1.5‚éØ‚éØ‚éØ‚éØ‚éØ‚éØ‚éØ‚éØ,‚àí1,‚àí0.5‚éØ‚éØ‚éØ‚éØ‚éØ‚éØ‚éØ‚éØ,0].

```python
#–ó–∞–¥–∞—á–∞ 2

t=int(input()) #–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –º–∞—Å—Å–∏–≤–æ–≤
assert 1<=t<=100, "–í–≤–µ–¥–µ–Ω–æ –Ω–µ–¥–æ–ø—É—Å—Ç–∏–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ"

for i in range(t):        
    print("–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:") 
    n=int(input()) #–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —á–∏—Å–µ–ª –≤ –∫–∞–∂–¥–æ–º –º–∞—Å—Å–∏–≤–µ
    assert 1<=n<=70, "–í–≤–µ–¥–µ–Ω–æ –Ω–µ–¥–æ–ø—É—Å—Ç–∏–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ"
    a=list(map(int,input().split())) #—á–∏—Å–ª–∞
    assert len(a)==n, "–í–≤–µ–¥–µ–Ω–æ –Ω–µ–¥–æ–ø—É—Å—Ç–∏–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ"
    
    t=0
    ans=0
    for i in range(0,n-1):
        t=0
        d={}
        for j in range(i+1,n):
            k=((a[j]-a[i])*1.0)/(i-j) #–ø—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫—É—é –ø—Ä–æ–≥—Ä–µ—Å—Å–∏—é
            if k not in d:
                d[k]=1
            else:
                d[k]+=1
            ans=max(ans,d[k])
            
    print("–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:")
    print(n-ans-1)
```

```python
#–ó–∞–¥–∞—á–∞ 3

from collections import Counter
t=int(input())
assert 1<=t<=1000, "–í–≤–µ–¥–µ–Ω–æ –Ω–µ–¥–æ–ø—É—Å—Ç–∏–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ"

for _ in range(t):
    print("–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:")
    n,l,r=list(map(int,input().split()))
    assert 2<=n<=2*105 and 0 <=l and r<=n and l+r == n and n%2 == 0, "–í–≤–µ–¥–µ–Ω–æ –Ω–µ–¥–æ–ø—É—Å—Ç–∏–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ"
    a=list(map(int,input().split()))
    assert len(a)==n, "–í–≤–µ–¥–µ–Ω–æ –Ω–µ–¥–æ–ø—É—Å—Ç–∏–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ"
    
    count=n//2 #–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–∞—Ä
    x=abs(l-r)//2 #–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–æ—Å–∫–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –±—É–¥—É—Ç —Å—Ç–∞–Ω–æ–≤–∏—Ç—å—Å—è –¥—Ä—É–≥–∏–º–∏ - –ø—Ä–∞–≤—ã–º–∏ –∏–ª–∏ –ª–µ–≤—ã–º–∏
    c=Counter(a[:l]) #–ª–µ–≤—ã–µ –Ω–æ—Å–∫–∏
    d=Counter(a[l:]) #–ø—Ä–∞–≤—ã–µ
    z=0
    if l>r: #–µ—Å–ª–∏ –ª–µ–≤—ã—Ö –±–æ–ª—å—à–µ, —Ç–æ –º–µ–Ω—è—é –∏—Ö –º–µ—Å—Ç–∞–º–∏ —Å –ø—Ä–∞–≤—ã–º–∏, —á—Ç–æ–±—ã –ø—Ä–∞–≤–∏–ª—å–Ω–æ –ø–æ—Å—á–∏—Ç–∞—Ç—å
        l,r=r,l
        c,d=d,c
    for i in c: #–∏–¥–µ–º –ø–æ –ª–µ–≤—ã–º –Ω–æ—Å–∫–∞–º
        m=min(c[i],d[i]) #—Å—Ä–∞–≤–Ω–∏–≤–∞–µ–º —Ç–æ, —Å–∫–æ–ª—å–∫–æ —Ä–∞–∑ –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è —Ü–≤–µ—Ç –≤ –ª–µ–≤–æ–π –∏ –ø—Ä–∞–≤–æ–π —á–∞—Å—Ç–∏, –≤—ã–±–∏—Ä–∞–µ–º –º–µ–Ω—å—à–µ–µ
        d[i]-=m #–Ω–∞ –≤—ã–±—Ä–∞–Ω–Ω–æ–µ –º–µ–Ω—å—à–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —É–º–µ–Ω—å—à–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç—Ç–æ–≥–æ —Ü–≤–µ—Ç–∞ –≤ –ø—Ä–∞–≤–æ–π —á–∞—Å—Ç–∏
        count-=m #–∏ —É–º–µ–Ω—å—à–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–∞—Ä, –¥–ª—è –∫–æ—Ç–æ—Ä—ã—Ö –Ω—É–∂–Ω—ã –±—É–¥—É—Ç –∏–∑–º–µ–Ω–µ–Ω–∏—è
    for i in d:
        z+=d[i]//2 #–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–∞–≤—ã—Ö –Ω–æ—Å–∫–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ —Å—Ç–∞–Ω—É—Ç –ª–µ–≤—ã–º–∏
        
    print("–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:") #–≤ –∏—Ç–æ–≥–µ —Å–∫–ª–∞–¥—ã–≤–∞–µ–º –∫–æ–ª-–≤–æ –∏–∑–º–µ–Ω–µ–Ω–Ω—ã—Ö –ø–∞—Ä —Å –º–∏–Ω–∏–º–∞–ª—å–Ω—ã–º –∫–æ–ª-–≤–æ–º –∏–∑–º–µ–Ω–µ–Ω–Ω—ã—Ö –Ω–æ—Å–∫–æ–≤
    print(count+(x-min(x,z)))
```

```python
#–ó–∞–¥–∞—á–∞ 4

n = int(input())
assert 1<=n<=2*105, "–í–≤–µ–¥–µ–Ω–æ –Ω–µ–¥–æ–ø—É—Å—Ç–∏–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ"
s = input()
assert len(s)==n, "–í–≤–µ–¥–µ–Ω–æ –Ω–µ–¥–æ–ø—É—Å—Ç–∏–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ"
t = input()
assert len(t)==n, "–í–≤–µ–¥–µ–Ω–æ –Ω–µ–¥–æ–ø—É—Å—Ç–∏–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ"

ab = [] 
ba =[]
for i in range(n): 
    if s[i]=='a' and t[i]=='b': #–≤ ab —Å–æ—Ö—Ä–∞–Ω—è–µ–º –ø–æ–∑–∏—Ü–∏–∏ i, —Ç–∞–∫–∏–µ, —á—Ç–æ s[i] = a –∏ t[i] = b
        ab.append(i+1)
    if s[i]=='b' and t[i]=='a': #–≤ ba –Ω–∞–æ–±–æ—Ä–æ—Ç
        ba.append(i+1)
if (len(ab)+len(ba))%2: #–µ—Å–ª–∏ —Ä–∞–∑–º–µ—Ä—ã ab u ba –ø–æ –º–æ–¥—É–ª—é 2 –Ω–µ —Ä–∞–≤–Ω—ã, —Ç–æ –æ—Ç–≤–µ—Ç–∞ –Ω–µ—Ç
    print(-1)           #–ø–æ—Ç–æ–º—É —á—Ç–æ —Å—É–º–º–∞—Ä–Ω—ã–µ –∫–æ–ª-–≤–∞ a –∏ b –≤ —Å—Ç—Ä–æ–∫–∞—Ö —Ç–æ–≥–¥–∞ –Ω–µ—á—ë—Ç–Ω—ã
else:
    ans=[]
    ans=[(ab[2*a],ab[2*a+1]) for a in range(len(ab)//2)] #–æ–ø—Ä–µ–¥–µ–ª—è–µ–º –ø–æ–∑–∏—Ü–∏–∏, –∫–æ—Ç–æ—Ä—ã–µ –Ω—É–∂–Ω—ã –¥–ª—è –æ–±–º–µ–Ω–∞ –±—É–∫–≤
    ans+=[(ba[2*a],ba[2*a+1]) for a in range(len(ba)//2)]
    if len(ab)%2: #–µ—Å–ª–∏ ab –Ω–µ—á–µ—Ç–Ω–∞, —Ç–æ —É—Ä–∞–≤–Ω–∏–≤–∞–µ–º —Å—Ç—Ä–æ–∫–∏
        ans+=[(ab[-1],ab[-1]),(ba[-1],ab[-1])]
    print(len(ans))
    print('\n'.join(f'{i[0]} {i[1]}'for i in ans))
```

–í –ø–µ—Ä–≤–æ–º –ø—Ä–∏–º–µ—Ä–µ 7 –º–æ–∂–µ—Ç –±—ã—Ç—å –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∞ –∫–∞–∫ 7=1+6, –≥–¥–µ 1 –∏ 6 —è–≤–ª—è—é—Ç—Å—è —Å–∏–ª—å–Ω—ã–º–∏. –¢–∞–∫ –∫–∞–∫ 7 –Ω–µ —è–≤–ª—è–µ—Ç—Å—è —Å–∏–ª—å–Ω—ã–º —á–∏—Å–ª–æ–º, –º—ã –∑–Ω–∞–µ–º, —á—Ç–æ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ ùëò –≤ —ç—Ç–æ–º —Å–ª—É—á–∞–µ ùëò=2.

–í–æ –≤—Ç–æ—Ä–æ–º –ø—Ä–∏–º–µ—Ä–µ –æ–¥–∏–Ω –∏–∑ —Å–ø–æ—Å–æ–±–æ–≤ —Ä–∞–∑–ª–æ–∂–∏—Ç—å 11 –≤ —Å—É–º–º—É —Ç—Ä–µ—Ö —Å–∏–ª—å–Ω—ã—Ö —á–∏—Å–µ–ª —Ç–∞–∫–æ–π: 11=1+4+6. –ú–æ–∂–Ω–æ –ø–æ–∫–∞–∑–∞—Ç—å, —á—Ç–æ –Ω–µ–ª—å–∑—è –ø—Ä–µ–¥—Å—Ç–∞–≤–∏—Ç—å 11 –≤ –≤–∏–¥–µ —Å—É–º–º—ã –¥–≤—É—Ö –∏–ª–∏ –º–µ–Ω–µ–µ —Å–∏–ª—å–Ω—ã—Ö —á–∏—Å–µ–ª.

–í —Ç—Ä–µ—Ç—å–µ–º –ø—Ä–∏–º–µ—Ä–µ 240 –º–æ–∂–µ—Ç –±—ã—Ç—å –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–æ –∫–∞–∫ 240=24+32+64+120. –û–±—Ä–∞—Ç–∏—Ç–µ –≤–Ω–∏–º–∞–Ω–∏–µ, —á—Ç–æ 240=120+120 –Ω–µ —è–≤–ª—è–µ—Ç—Å—è –∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–º –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ–º, —Ç. –∫. —Å–∏–ª—å–Ω—ã–µ —á–∏—Å–ª–∞ –¥–æ–ª–∂–Ω—ã –±—ã—Ç—å —Ä–∞–∑–ª–∏—á–Ω—ã.

–í —á–µ—Ç–≤–µ—Ä—Ç–æ–º –ø—Ä–∏–º–µ—Ä–µ 17179869184=234, –ø–æ—ç—Ç–æ–º—É 17179869184 —è–≤–ª—è–µ—Ç—Å—è —Å–∏–ª—å–Ω—ã–º —á–∏—Å–ª–æ–º, –∏ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ ùëò –≤ —ç—Ç–æ–º —Å–ª—É—á–∞–µ ùëò=1.

```python
#–ó–∞–¥–∞—á–∞ 10

fact = [1] * 15
subs = []
 
for i in range(1, 15):
    fact[i] = fact[i - 1] * (i + 1)

for i in range(1 << 15): #<< - –ø–æ–±–∏—Ç–æ–≤—ã–π —Å–¥–≤–∏–≥ –≤–ª–µ–≤–æ
    sub, s = 0, 0
    for j in range(15): #–ø–µ—Ä–µ–±–∏—Ä–∞–µ–º –≤—Å–µ –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç–∏ –≤–∫–ª—é—á–µ–Ω–∏—è –∏–ª–∏ –Ω–µ–≤–∫–ª—é—á–µ–Ω–∏—è –∫–∞–∂–¥–æ–≥–æ —Ñ–∞–∫—Ç–æ—Ä–∏–∞–ª–∞ –¥–æ 14
        if i & (1 << j): #–¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∏–∑ –Ω–∏—Ö –≤—ã—á–∏—Å–ª—è–µ–º –∫–æ–ª-–≤–æ —á–∏—Å–µ–ª, –∏—Å–ø–æ–ª—å–∑—É–µ–º—ã—Ö –≤ —Å—É–º–º–µ
            sub += 1
            s += fact[j]
    subs.append((sub, s)) #–¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —Å–∏–ª—å–Ω—ã—Ö —á–∏—Å–µ–ª, –∏—Å–ø–æ–ª—å–∑—É–µ–º—ã—Ö –≤ —Å—É–º–º–µ

t = int(input())
assert 1<=t<=100, "–í–≤–µ–¥–µ–Ω–æ –Ω–µ–¥–æ–ø—É—Å—Ç–∏–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ"
for _ in range(t):
    print("–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:")
    n = int(input())
    assert 1<=n<=1012, "–í–≤–µ–¥–µ–Ω–æ –Ω–µ–¥–æ–ø—É—Å—Ç–∏–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ"
 
    ans = bin(n).count('1') #–∏—Å–ø–æ–ª—å–∑—É–µ–º –¥–≤–æ–∏—á–Ω–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ n, –∫–æ–ª-–≤–æ —á–∏—Å–µ–ª –±—É–¥–µ—Ç —Ä–∞–≤–Ω—ã–º 1 –≤ –¥–≤–æ–∏—á–Ω–æ–º –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–∏
 
    for i in subs: 
        if i[1] <= n: 
            ans = min(ans, i[0] + bin(n - i[1]).count('1')) #–≤—ã—á–∏—Å–ª—è–µ–º –∫–æ–ª-–≤–æ —á–∏—Å–µ–ª, –Ω–∞—Ö–æ–¥–∏–º –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ
 
    print("–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:")
    print(ans)
```
